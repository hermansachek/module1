# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

```python
home_wins = df_regular_season.groupby("home_team")["home_team_win"].sum()
away_wins = df_regular_season.groupby("away_team")["home_team_loss"].sum()
total_wins = home_wins.add(away_wins, fill_value = 0).sort_values(ascending=False) 
total_wins.head(1)
```

Победы домашней команды мы можем определить по столбцу "home_team_win", а победу гостевой по столбцу "home_team_loss", так как поражение 
домашней команды означает победу готсевой команды. В total_wins передаем сумму побед в домашних и гостивых матчах, сортируя по убывания и оставляя 
только самую верхнюю строку, которая будет выводить название команды с наибольшим кол-вом побед с сезоне.

# 20. Правда ли что наибольшее количество ранов происходит в холодную погоду? (Холодной погодой считается погода ниже 0 градусов)
